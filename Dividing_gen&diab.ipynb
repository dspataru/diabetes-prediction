After checking the count of Null values for each colums, we realised that the a few columns have question specific to the patients who have diabetes therefore they all have same number of missing values (questions that were not asked from the non-diabetic people). Therefore, a two separate dataframes are created"

- 1. Containing the general information <br>
Will be used to categorize if the person has diabetes or not
Columns specific to Diabetes are dropped new dataframe "gen_info_df"

- 2. Specific to Diabetic patients <br>
Will be used to categorize the type of diabetes


**These two dataframes can be joined on the "ID Column"**


```python
# Creating a separate df with columns containg values for data-specifc to people having diabetes
diabetes_df = data_df[["ID", "GENDER", "AGE", "BMI", "DIABETES", "DIABTYPE", "PREDIABETIC", "BLD_SUG_TST", "INSULIN_Y/N", 
                      "A-one-C_test", "EYEEXAM1", "DIABEYE1", "DIAB_MNGMT", "FEETSORE", "PERSONAL_DOC", "HRT_ATTACK", 
                     "HRT_DISEASE", "STROKE", "ARTHRITIS"]] 

diabetes_df.shape
```

```python
#Removing columns specific to the people who have diabetes and creating a new dataframe with all the general info
# Dropping 9 columns-specifc to diabetes
gen_info_df= data_df.drop(columns = ["DIABTYPE", "PREDIABETIC", "BLD_SUG_TST", "INSULIN_Y/N", 
                      "A-one-C_test", "EYEEXAM1", "DIABEYE1", "DIAB_MNGMT", "FEETSORE"])
gen_info_df.shape
```

```python
#Finding the null values for each column
gen_info_df.isnull().sum()
```

```python
#Drop all the NaN values
gen_info_df = gen_info_df.dropna(how='any')
```

```python
gen_info_df.shape
```

```python
#Finding the null values for each column
gen_info_df.isnull().sum()
```

```python
#Save the data cleaned data into a .csv file
gen_info_df.to_csv('final_gen_info.csv', index=False)
```

```python
#For age the Nan values were replaced with mean of entire column
diab_df['AGE'].fillna((diab_df['AGE'].mean()), inplace=True)
```

```python
#For BMI the Nan values were replaced randomly selected category based on the distribution of existing values.
import random
categories = diab_df['BMI'].dropna().unique()
diab_df['BMI'].fillna(random.choice(categories), inplace=True)
```

```python
#Finding the null values for each column
diab_df.isnull().sum()
```

```python
#Replaced all other the "NaN" values with 0
diab_df= diab_df.fillna(0)
```

```python
# iterate over the list again to confirm that the changes are implemented successfully
column = diab_df.columns
for col in column:
    print(col, ':', str(diab_df[col].unique()))
```
