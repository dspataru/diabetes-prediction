## Feature Engineering

- Converting categorical value to category
- Changing the value of Don't Know/refused to NaN values

## Categorical columns :
'SEXVAR', '_BMI5CAT', '_RACE1', '_EDUCAG', '_SMOKER3'

```python
#Substituting the value of 'SEXVAR'- 1: MALE, 2: FEMALE
Gender_dict = {1:"MALE",
            2: "FEMALE"}
imp_df['SEXVAR'] = imp_df['SEXVAR'].apply(lambda x: Gender_dict[x])
```

```python
#Substituting the value of '_BMI5CAT'- 1: Underweight, 2 : Normal_Weight, 3: Over_Weight, 4: Obese
#NaN_value was creating an error, therfore used map
BMI_dict = {1: "Underweight",
            2 : "Normal_Weight", 
            3: "Over_Weight", 
            4: "Obese"}
imp_df['_BMI5CAT'] = imp_df['_BMI5CAT'].map(BMI_dict)
```

```python
#Substituting the value of '_RACE1'- 1: "White",2: "Black",3: "Indian-Alaskan_Native", 4: "Asian",5: "Hawaiian",
#7: "Multiracial",8: "Hispanic",9:  np.nan 

Race_dict = {1: "White",
             2: "Black",
             3: "Indian-Alaskan_Native", 
             4: "Asian",
             5: "Hawaiian",
             7: "Multiracial",
             8: "Hispanic",
             9:  np.nan    
}

imp_df['_RACE1'] = imp_df['_RACE1'].map(Race_dict)
```

```python
#Substituting the value of '_EDUCAG'- 1: "Grad_HS_N",2: "Grad_HS_Y", 3: "College_N", 4: "College_Y",9:  np.nan
Edu_dict = {1: "Grad_HS_N",
            2: "Grad_HS_Y", 
            3: "College_N", 
            4: "College_Y",
            9:  np.nan
            }

imp_df['_EDUCAG'] = imp_df['_EDUCAG'].map(Edu_dict)
```

```python
#Substituting the value of '_SMOKER3' - 1: "Smok_daily_Y", 2: "Smok_daily_N", 3: "Prev_Smoker", 4: "Never", 9:  np.nan
Smok_dict = {1: "Smok_daily_Y", 
             2: "Smok_daily_N", 
             3: "Prev_Smoker", 
             4: "Never", 
             9:  np.nan,
            }

imp_df['_SMOKER3'] = imp_df['_SMOKER3'].map(Smok_dict)
```

## Numerical Columns

- Different columns have different features with 'Dont know/Not sure' value and "No/Never" value , this section works around replacing those values to either NaN or 0 respectively
- Evaluating the "Height feature (HEIGHT3)"

```python
#for column "_INCOMG1", value 9 represents 'Dont know/Not sure' value, but value 7 represents $200,000 or more
#therefore only 9 has to be replaced by value NaN
imp_df["_INCOMG1"].replace({9 : np.nan},inplace=True)
```

#### for column "HEIGHT3" 

- 200 - 711: Height (ft/inches)<br>
--Notes: 0 _ / _ _ = feet / inches-- i.e. the first digit is feet, the second and third digits are inches. So 509 is 5' 9''

- 9061 - 9998	Height (meters/centimeters)<br>
Notes: The initial 9  indicates this was a metric value. Height in m/cm (9_|_ _)

- 7777:	Dont know/Not sure	

- 9999: Refused

```python
for indx, val in imp_df["HEIGHT3"].iteritems(): 

    h = str(val)
    if val < 712:
        feet = float(h[0])
        inches = float(h[1:])
        H_inches = inches + (feet*12)
        imp_df.loc[indx, "HEIGHT"] = H_inches
           
    elif val<9999 and val>9060:
        meter = float(h[1])
        cm = float(h[2:])
        H_meter = meter+(cm*0.01)
        H_inches = H_meter*39.37
        imp_df.loc[indx, "HEIGHT"] = H_inches
                 
    
    else:
        val== 7777 or val== 9999
        imp_df.loc[indx, "HEIGHT"]  = np.nan
    
    
```

```python
#for column "_AGEG5YR", value 14 represents 'Dont know/Not sure' value
imp_df["_AGEG5YR"].replace({14 : np.nan},inplace=True)
```

```python
#Columns with 'Dont know/Not sure' or 'Refused' value as 7 or 9 
columns = ["DIABETE4", "PREDIAB2", "DIABTYPE", "_TOTINDA", "PERSDOC3", "CHECKUP1", "PDIABTS1", "INSULIN1", "EYEEXAM1",
          "DIABEYE1", "DIABEDU1", "FEETSORE", "CVDINFR4", "CVDCRHD4", "CVDSTRK3", "HAVARTH4","DIFFWALK"]

#Iterate through each column and replace the value of 7 or 9 by NaN value
for column in columns:
    imp_df[column].replace({7 : np.nan}, inplace=True)
    imp_df[column].replace({9 : np.nan}, inplace=True)
```

```python
# iterate over the list again to confirm that the changes are implemented successfully
for column in columns:
    print(column, ':', str(imp_df[column].unique()))
```

```python
columns = ["CHECKUP1", "PDIABTS1", "EYEEXAM1","DIABEYE1", "DIABEDU1"]
```

```python
# iterate over the list again to confirm that the changes are implemented successfully
for column in columns_3:
    print(column, ':', str(imp_df[column].unique()))
```

```python
#Column where value 8 represents "Never" replacing the value of 8 as 0 
columns_4 = ["CHECKUP1", "PDIABTS1", "EYEEXAM1","DIABEYE1", "DIABEDU1"]
#Iterate through each column and replace the value of 7 or 9 by NaN value
for column in columns_4:
    imp_df[column].replace({8 : 0}, inplace=True)
```

```python
#Columns with 'Dont know/Not sure' or 'Refused' value as 77 or 99 
columns_2 = ["CHKHEMO3", "PHYSHLTH", "MENTHLTH", "SLEPTIM1", "PRIMINSR"]

#Iterate through each column and replace the value of 77 or 99 by NaN value
for column in columns_2:
    imp_df[column].replace({77 : np.nan}, inplace=True)
    imp_df[column].replace({99 : np.nan},  inplace=True)
```

```python
# iterate over the list again to confirm that the changes are implemented successfully
for column in columns_2:
    print(column, ':', str(imp_df[column].unique()))
```

```python
#Columns in column_2 have a value 88 with "None" or "Never" converting this value to 0
columns_2 = ["CHKHEMO3", "PHYSHLTH", "MENTHLTH", "SLEPTIM1", "PRIMINSR"]

#Iterate through each column and replace the value of 77 or 99 by NaN value
for column in columns_2:
    imp_df[column].replace({88 : 0}, inplace=True)
```
