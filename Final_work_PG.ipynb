# 5.2 Under and Over Sampling

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import StandardScaler
```

## Naming the Target and Features and Split training/test datasets

```python
# Separate features and target variable
target_variable = diabetes_df['Diabetes']
features = diabetes_df.drop(columns='Diabetes')

# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(features, target_variable, random_state=42, stratify=target_variable)

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Train a logistic regression model over and under sampling

```python
# Oversampling using SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

# Undersampling using RandomUnderSampler
undersampler = RandomUnderSampler(random_state=42)
X_train_resampled, y_train_resampled = undersampler.fit_resample(X_train_scaled, y_train)

# Logistic Regression with SMOTE
lr_smote = LogisticRegression(random_state=10)
lr_smote.fit(X_train_smote, y_train_smote)
y_pred_smote = lr_smote.predict(X_test_scaled)
print('Logistic Regression with SMOTE:\n', classification_report(y_test, y_pred_smote))

# Logistic Regression with RandomUnderSampler
lr_resampled = LogisticRegression(random_state=10)
lr_resampled.fit(X_train_resampled, y_train_resampled)
y_pred_resampled = lr_resampled.predict(X_test_scaled)
print('Logistic Regression with RandomUnderSampler:\n', classification_report(y_test, y_pred_resampled))
```
