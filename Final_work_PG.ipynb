```python
# For the machine learning model, our focus is to determine whether someone has diabetes or is trending towards diabetes.
# The label encoding is as follows:
# 0: No diabetes or only reported during pregnancy for females ('no')
# 1: Pre-diabetes or borderline diabetes ('no')
# 2: Diabetes ('yes')
# We will exclude data points marked as "Don't know/Not Sure" and "Refused."

data_for_model_['Diabetes'] = data_for_model_['Diabetes'].replace({1:2, 2:0, 3:0, 4:1})
data_for_model_ = data_for_model_[data_for_model_.Diabetes != 7]
data_for_model_ = data_for_model_[data_for_model_.Diabetes != 9]
data_for_model_.Diabetes.value_counts()
```

```python
from collections import Counter
from imblearn.combine import SMOTETomek, SMOTEENN
from imblearn.over_sampling import ADASYN, SMOTE
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Generate a synthetic imbalanced dataset
X, y = make_classification(n_classes=2, class_sep=2,
                           weights=[0.1, 0.9], n_informative=3,
                           n_redundant=1, flip_y=0, n_features=20,
                           n_clusters_per_class=1, n_samples=1000,
                           random_state=10)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

# Scale the training and testing sets
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Print the class distribution of the original training dataset
counter = Counter(y_train)
print('Before oversampling:', counter)

# Oversample the training dataset using four different algorithms
smenn = SMOTEENN()
smt = SMOTE()
ada = ADASYN(sampling_strategy=0.5, random_state=130)
smtom = SMOTETomek(random_state=139)

X_train_smenn, y_train_smenn = smenn.fit_resample(X_train_scaled, y_train) # SMOTE + ENN algorithm
X_train_sm, y_train_sm = smt.fit_resample(X_train_scaled, y_train) # SMOTE algorithm
X_train_ada, y_train_ada = ada.fit_resample(X_train_scaled, y_train) # ADASYN algorithm
X_train_smtom, y_train_smtom = smtom.fit_resample(X_train_scaled, y_train) # SMOTE + Tomek algorithm

# Print the class distribution of the oversampled datasets
print('After SMOTE:', Counter(y_train_sm))
print('After SMOTEENN:', Counter(y_train_smenn))
print('After ADASYN:', Counter(y_train_ada))
print('After SMOTETomek:', Counter(y_train_smtom))
```

## Train a logistic regression model on each oversampled dataset and evaluate its performance on the testing set. 

```python
# Train a logistic regression model on each oversampled dataset and evaluate its performance on the testing set
lr_smenn = LogisticRegression(random_state=10)
lr_smenn.fit(X_train_smenn, y_train_smenn)
y_pred_smenn = lr_smenn.predict(X_test_scaled)
print('SMOTE + ENN:\n', classification_report(y_test, y_pred_smenn))

lr_sm = LogisticRegression(random_state=10)
lr_sm.fit(X_train_sm, y_train_sm)
y_pred_sm = lr_sm.predict(X_test_scaled)
print('SMOTE:\n', classification_report(y_test, y_pred_sm))

lr_ada = LogisticRegression(random_state=10)
lr_ada.fit(X_train_ada, y_train_ada)
y_pred_ada = lr_ada.predict(X_test_scaled)
print('ADASYN:\n', classification_report(y_test, y_pred_ada))

lr_smtom = LogisticRegression(random_state=10)
lr_smtom.fit(X_train_smtom, y_train_smtom)
y_pred_smtom = lr_smtom.predict(X_test_scaled)
print('SMOTE + Tomek:\n', classification_report(y_test, y_pred_smtom))
```

```python
from sklearn.model_selection import cross_val_score
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN

rus = RandomUnderSampler(random_state=130)
X_train_under, y_train_under = rus.fit_resample(X_train, y_train)

ros = RandomOverSampler(random_state=130)
X_train_over, y_train_over = ros.fit_resample(X_train, y_train)

smote = SMOTE(random_state=130)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

smote_enn = SMOTEENN(random_state=130)
X_train_smenn, y_train_smenn = smote_enn.fit_resample(X_train, y_train)

# Define the resampled datasets
datasets = {'Original': (X_train, y_train),
            'Undersampled': (X_train_under, y_train_under),
            'Oversampled': (X_train_over, y_train_over),
            'SMOTE': (X_train_smote, y_train_smote),
            'SMOTE + ENN': (X_train_smenn, y_train_smenn)}

# Define the models
models = {'Logistic Regression': LogisticRegression(random_state=130),
          'Random Forest': RandomForestClassifier(random_state=130)}

# Evaluate the models on each dataset
for dataset_name, (X, y) in datasets.items():
    print('Dataset:', dataset_name)
    for model_name, model in models.items():
        scores = cross_val_score(model, X, y, cv=5, scoring='f1')
        print(model_name, 'F1-score:', scores.mean())
    print()
```

## Making the ROC curve for Over and Under Sampling

```python
# Generate a synthetic imbalanced dataset
X, y = make_classification(n_classes=2, class_sep=2,
                           weights=[0.1, 0.9], n_informative=3,
                           n_redundant=1, flip_y=0, n_features=20,
                           n_clusters_per_class=1, n_samples=1000,
                           random_state=10)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

# Scale the training and testing sets
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Oversample the training dataset using SMOTE
smote = SMOTE(random_state=10)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)

# Train a logistic regression model on the oversampled dataset
lr_over = LogisticRegression(random_state=10)
lr_over.fit(X_train_resampled, y_train_resampled)

# Predict the probabilities of the target variable for the test data using the trained model
y_prob_over = lr_over.predict_proba(X_test_scaled)[:, 1]

# Calculate the False Positive Rate (FPR) and True Positive Rate (TPR) for different threshold values
fpr_over, tpr_over, thresholds_over = roc_curve(y_test, y_prob_over)

# Calculate the area under the ROC curve
roc_auc_over = auc(fpr_over, tpr_over)

# Undersample the training dataset using RandomUnderSampler
rus = RandomUnderSampler(random_state=10)
X_train_resampled, y_train_resampled = rus.fit_resample(X_train_scaled, y_train)

# Train a logistic regression model on the undersampled dataset
lr_under = LogisticRegression(random_state=10)
lr_under.fit(X_train_resampled, y_train_resampled)

# Predict the probabilities of the target variable for the test data using the trained model
y_prob_under = lr_under.predict_proba(X_test_scaled)[:, 1]

# Calculate the False Positive Rate (FPR) and True Positive Rate (TPR) for different threshold values
fpr_under, tpr_under, thresholds_under = roc_curve(y_test, y_prob_under)

# Calculate the area under the ROC curve
roc_auc_under = auc(fpr_under, tpr_under)

# Plot the ROC curves for the oversampled and undersampled models
plt.plot(fpr_over, tpr_over, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f) - Oversampled' % roc_auc_over)
plt.plot(fpr_under, tpr_under, color='blue', lw=2, label='ROC curve (AUC = %0.2f) - Undersampled' % roc_auc_under)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic (ROC) curve')
plt.legend(loc="lower right")
plt.show()
```
