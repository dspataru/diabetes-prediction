# 2. Cleaning data to make it more suitable for ML

To accomplish this task, every column in the dataset underwent a thorough review in reference to the associated codebook. The codebook provides information about the nature of each column and elucidates the specific meanings and correspondences of the values within each column.

1. Diabetes	

```python
# For the machine learning model, our focus is to determine whether someone has diabetes or is trending towards diabetes.
# The label encoding is as follows:
# 0: No diabetes or only reported during pregnancy for females ('no')
# 1: Pre-diabetes or borderline diabetes ('no')
# 2: Diabetes ('yes')
# We will exclude data points marked as "Don't know/Not Sure" and "Refused."

data_for_model_['Diabetes'] = data_for_model_['Diabetes'].replace({1:2, 2:0, 3:0, 4:1})
data_for_model_ = data_for_model_[data_for_model_.Diabetes != 7]
data_for_model_ = data_for_model_[data_for_model_.Diabetes != 9]
data_for_model_.Diabetes.value_counts()
```

2. Smokers

```python
# We will remove the Don’t know/Refused and keep only the yes ('1') and no ('2')
# Replace 2 with 0
data_for_model_['Smoker'] = data_for_model_['Smoker'].replace({1:1, 2:1, 3:1, 4:0})
data_for_model_ = data_for_model_[data_for_model_.Smoker != 9]
data_for_model_.Smoker.value_counts()
```

3. CHD_1 - Chronic Health Conditions

```python
# We will remove the Don’t know/Refused and keep only the yes ('1') and no ('2')
# Replace 2 with 0
data_for_model_['CHD_1'] = data_for_model_['CHD_1'].replace({2:0})
data_for_model_ = data_for_model_[data_for_model_.CHD_1 != 7]
data_for_model_ = data_for_model_[data_for_model_.CHD_1 != 9]
data_for_model_.CHD_1.value_counts()
```

# 5.2 Under and Over Sampling

```python
target_variable = new_diabetes_df['Diabetes']
features = new_diabetes_df.drop(columns = 'Diabetes')
```

```python
# split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(features, target_variable, random_state=42, stratify=target_variable)

# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from imblearn.combine import SMOTETomek
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import ADASYN
from collections import Counter

counter = Counter(y_train)
print('Before', counter)

# oversampling the train dataset using SMOTE + ENN + ADASYN
smenn = SMOTEENN()
smt = SMOTE()
ada = ADASYN(random_state=130)
smtom = SMOTETomek(random_state=139)

X_train_smenn, y_train_smenn = smenn.fit_resample(X_train_scaled, y_train) # SMOTE + ENN algorithm
X_train_sm, y_train_sm = smt.fit_resample(X_train_scaled, y_train) # SMOTE algorithm
X_train_ada, y_train_ada = ada.fit_resample(X_train_scaled, y_train) # ADASYN algorithm
X_train_smtom, y_train_smtom = smtom.fit_resample(X_train_scaled, y_train) # SMOTE + Tomek algorithm

print('After SMOTE:', y_train_sm)
print('After SMOTEENN:', y_train_smenn)
print('After ADASYN:', y_train_ada)
print('After SMOTETomek:', y_train_smtom)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score, f1_score

# Logistic regression
lr = LogisticRegression(random_state=130)
lr.fit(X_train_smenn, y_train_smenn)

y_pred_lr = lr.predict(X_test_scaled)

print('Logistic Regression:')
print('Accuracy:', accuracy_score(y_test, y_pred_lr))
print('Precision:', precision_score(y_test, y_pred_lr))
print('Recall:', recall_score(y_test, y_pred_lr))
print('F1-score:', f1_score(y_test, y_pred_lr))
```

```python
# Random forest
rf = RandomForestClassifier(random_state=130)
rf.fit(X_train_smenn, y_train_smenn)

y_pred_rf = rf.predict(X_test_scaled)

print('Random Forest:')
print('Accuracy:', accuracy_score(y_test, y_pred_rf))
print('Precision:', precision_score(y_test, y_pred_rf))
print('Recall:', recall_score(y_test, y_pred_rf))
print('F1-score:', f1_score(y_test, y_pred_rf))
```

```python
from sklearn.model_selection import cross_val_score
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN

rus = RandomUnderSampler(random_state=130)
X_train_under, y_train_under = rus.fit_resample(X_train, y_train)

ros = RandomOverSampler(random_state=130)
X_train_over, y_train_over = ros.fit_resample(X_train, y_train)

smote = SMOTE(random_state=130)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

smote_enn = SMOTEENN(random_state=130)
X_train_smenn, y_train_smenn = smote_enn.fit_resample(X_train, y_train)

# Define the resampled datasets
datasets = {'Original': (X_train, y_train),
            'Undersampled': (X_train_under, y_train_under),
            'Oversampled': (X_train_over, y_train_over),
            'SMOTE': (X_train_smote, y_train_smote),
            'SMOTE + ENN': (X_train_smenn, y_train_smenn)}

# Define the models
models = {'Logistic Regression': LogisticRegression(random_state=130),
          'Random Forest': RandomForestClassifier(random_state=130)}

# Evaluate the models on each dataset
for dataset_name, (X, y) in datasets.items():
    print('Dataset:', dataset_name)
    for model_name, model in models.items():
        scores = cross_val_score(model, X, y, cv=5, scoring='f1')
        print(model_name, 'F1-score:', scores.mean())
    print()
```
